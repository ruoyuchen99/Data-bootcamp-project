```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
pd.set_option('display.max_rows', None)
mpl.style.use('seaborn')
```

# Income based visualization
# 1. WGI data

## Low income countries

```python
# import data for low income countries
low = pd.read_csv('Low_inc_WGI.csv')

#cleaning and reshaping data for low income countries
low=low.rename(columns={'2004 [YR2004]':2004,'2005 [YR2005]':2005,'2006 [YR2006]':2006,'2007 [YR2007]':2007,'2008 [YR2008]':2008,\
                        '2009 [YR2009]':2009,'2010 [YR2010]':2010,'2011 [YR2011]':2011,'2012 [YR2012]':2012,\
                        '2013 [YR2013]':2013,'2014 [YR2014]':2014,'2015 [YR2015]':2015,\
                        '2016 [YR2016]':2016,'2017 [YR2017]':2017,'2018 [YR2018]':2018})

for i in low.columns[4:]:
    low[i]=pd.to_numeric(low[i],errors='coerce')
    
low=low.groupby('Series Name').mean()

low=low.transpose()
low=low.rename(columns={'Control of Corruption: Estimate':'corruption','Government Effectiveness: Estimate':'effectiveness',
                  'Political Stability and Absence of Violence/Terrorism: Estimate':'stability','Regulatory Quality: Estimate':'regulation',
                  'Rule of Law: Estimate':'law','Voice and Accountability: Estimate':'voice','Fixed broadband subscriptions (per 100 people)':'broadband'})

# plot 6 WGI indicators for low income countries over time
fig, ax = plt.subplots()
low.plot(ax=ax,figsize=(10,8))
ax.set_title('Low income WGI',size=24,fontweight='bold')
ax.set_xlabel('Year',size=16)

ax.legend(bbox_to_anchor=(1,1),prop={"size":15})
```
